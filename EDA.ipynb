<h3><code>playlists</code> field</h3>
<p>This is an array that typically contains 1,000 playlists. Each playlist is a dictionary that contains the following fields:</p>
<ul>
<li><strong><em>pid</em></strong> - integer - playlist id - the MPD ID of this playlist. This is an integer between 0 and 999,999.</li>
<li class=""><strong><em>name</em></strong> - string - the name of the playlist </li>
<li><strong><em>description</em></strong> - optional string - if present, the description given to the playlist.  Note that user-provided playlist descrptions are a relatively new feature of Spotify, so most playlists do not have descriptions.</li>
<li><strong><em>modified_at</em></strong> - seconds - timestamp (in seconds since the epoch) when this playlist was last updated. Times are rounded to midnight GMT of the date when the playlist was last updated.</li>
<li><strong><em>num_artists</em></strong> - the total number of unique artists for the tracks in the playlist.</li>
<li><strong><em>num_albums</em></strong> - the number of unique albums for the tracks in the playlist</li>
<li><strong><em>num_tracks</em></strong> - the number of tracks in the playlist</li>
<li><strong><em>num_followers</em></strong> - the number of followers this playlist had at the time the MPD was created. (Note that the follower count does not including the playlist creator)</li>
<li><strong><em>num_edits</em></strong> - the number of separate editing sessions. Tracks added in a two hour window are considered to be added in a single editing session.</li>
<li><strong><em>duration_ms</em></strong> - the total duration of all the tracks in the playlist (in milliseconds)</li>
<li><strong><em>collaborative</em></strong> -  boolean - if true, the playlist is a collaborative playlist. Multiple users may contribute tracks to a collaborative playlist.</li>
<li><strong><em>tracks</em></strong> - an array of information about each track in the playlist. Each element in the array is a dictionary with the following fields:

<ul>
<li><strong><em>track_name</em></strong> - the name of the track</li>
<li><strong><em>track_uri</em></strong> - the Spotify URI of the track</li>
<li><strong><em>album_name</em></strong> - the name of the track's album</li>
<li><strong><em>album_uri</em></strong> - the Spotify URI of the album</li>
<li><strong><em>artist_name</em></strong> - the name of the track's primary artist</li>
<li><strong><em>artist_uri</em></strong> - the Spotify URI of track's primary artist</li>
<li><strong><em>duration_ms</em></strong> - the duration of the track in milliseconds</li>
<li><strong><em>pos</em></strong> - the position of the track in the playlist (zero-based)</li>
</ul></li>
</ul>

```python
pid = [playlist['pid'] for playlist in  f['playlists'] for track in playlist['tracks']]
track_uri = [track['track_uri'] for playlist in  f['playlists'] for track in playlist['tracks']]
```

```python
df_2 = pd.DataFrame(data=[pid,track_uri]).T
```

```python
df_2.columns = ['pid','track_uri']
```

```python
df_2.head()
```

```python
occurrance_table = df_2.pivot_table(index=['pid'],columns='track_uri',aggfunc=lambda x: 1.0, fill_value=0)
```

```python
occurrance_table.T.head()
```

```python
from scipy.sparse import csr_matrix
```

```python
## condensing sparse matrix
## original numpy dot product causes takes a very long time to run

occurance_matrix = csr_matrix(occurrance_table)
```

```python
occurance_matrix.toarray().shape
```

```python
## transposition is so that dot product matrix will be track_uri rows vs track_uri cols
## user is taken out the matrix
## we have the numerator for sim(A,B) calculations

item_subset_occurance = (occurance_matrix.T*occurance_matrix).toarray()
```

```python
item_subset_occurance
```

```python
## total occurance per track

occurance_by_track = np.sum(occurance_matrix,axis=0,dtype=np.float)
```

```python
occurance_by_track.shape
```

```python
occurance_by_track = occurance_by_track.reshape(34443,1)
```

```python
occurance_by_track
```

```python
## Calculating the denominator for sim(A,B) calculations

item_union_occurance = (occurance_by_track+occurance_by_track.T)-item_subset_occurance
```

```python
similiarity_matrix = item_subset_occurance/item_union_occurance
```

```python
similiarity_matrix
```

```python
similarity_df = pd.DataFrame(similiarity_matrix,index = occurrance_table.columns, columns = occurrance_table.columns)
```
