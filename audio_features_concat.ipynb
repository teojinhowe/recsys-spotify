# Notebook to concatenate audio_features slices

```python
import pandas as pd
import numpy
import pickle
```

```python
audio_features = pickle.load(open('poc_song_audio_features_10000.sav','rb'))
```

```python
audio_features = dict()

i=10000
while i <=80000:
    print 'opening ' + 'poc_song_audio_features_{}.sav'.format(i)
    a = pickle.load(open('poc_song_audio_features_{}.sav'.format(i),'rb'))
    audio_features.update(a)
    print 'dict updated'
    i += 10000
```

```python
len(audio_features)
```

```python
df = pd.DataFrame(audio_features)
```

```python
df = df.T
```

```python
df.head()
```

```python
df.drop(labels=['analysis_url','id','track_href','type','uri'],axis=1,inplace=True)
```

```python
df.head()
```

```python
pickle.dump(df, open('poc_audio_features_new.sav','wb'))
```
